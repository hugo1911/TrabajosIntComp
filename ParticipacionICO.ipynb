```python
from openai import OpenAI

SYSTEM_MESSAGE = "You are a chatbot. You will have a conversation with a user. Be friendly and concise"

URL   = "https://generativelanguage.googleapis.com/v1beta/openai/"
KEY   = "AIzaSyAr5du5dJDsJU8NQvI36xxVU2odthRXEcE"
MODEL = "gemini-flash-latest"

client = OpenAI(
    base_url=URL,
    api_key=KEY,
)

print(f"que rollo andas hablando con {MODEL} en la url {URL}\n")

#Esto es lo que hace que guarde las cosas en el chat https://developers.openai.com/api/docs/guides/conversation-state$0
conversation_history = [
    {'role': 'system', 'content': SYSTEM_MESSAGE}
]

while True:
    message = input("> ")

    conversation_history.append({'role': 'user', 'content': message})

    response = client.chat.completions.create(
        model=MODEL,
        messages=conversation_history,
    )

    assistant_reply = response.choices[0].message.content

    conversation_history.append({'role': 'assistant', 'content': assistant_reply})

    print(f"\nAssistant: {assistant_reply}\n")
```
